```python
import tensorflow as tf
tf.__version__
```

```python
## version2에서는 Session이 없어졌음
a = tf.constant(10)
b = tf.constant(32)
c = tf.add(a,b)
print(c)
```

```python
matrix1 = tf.constant([[3,2]])
matrix2 = tf.constant(([2],[2]))

output = tf.matmul(matrix1, matrix2)
print(output)
```

```python
state = tf.Variable(0, name='counter')
one = tf.constant(1)

for _ in range(3):
    state = state + one
    print(state)
```

```python
@tf.function
def multiply(input1, input2):
    return input1 * input2

result = multiply(7,2)
print(result)
```
